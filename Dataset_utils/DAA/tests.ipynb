```python
import pandas as pd

# Load the CSV files
file1 = '/home/bas06400/daa/activities_3s/kinect_color/midlevel.chunks_90.split_0.val.csv'
file2 = '/home/bas06400/daa/activities_3s/inner_mirror/midlevel.chunks_90.split_0.val.csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Assuming both CSVs have the same number of rows and the same order of rows
# Calculate the differences in frame_start and frame_end between the two files
frame_start_diff = df2['frame_start'] - df1['frame_start']
frame_end_diff = df2['frame_end'] - df1['frame_end']

# Check if there's a consistent difference
consistent_start_diff = frame_start_diff.unique()
consistent_end_diff = frame_end_diff.unique()

print("Unique differences in frame_start:", consistent_start_diff)
print("Unique differences in frame_end:", consistent_end_diff)

# If you suspect the second file has frame_end values consistently higher by a certain value (e.g., +1),
# you can specifically check for that
specific_diff = 1  # for example, expecting a +1 difference
consistent_specific_end_diff = (df2['frame_end'] - df1['frame_end'] == specific_diff).all()

print("Is frame_end consistently higher by", specific_diff, "in the second file?", consistent_specific_end_diff)
```

```python
import numpy as np
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Double the frame_end from the first file and add 1
modified_frame_end_df1 = df1['frame_start']* 2 + 1

# Calculate the difference between modified frame_end from file1 and original frame_end from file2
diff = modified_frame_end_df1 - df2['frame_start']

# Calculate the average difference
average_diff = diff.abs().mean()

print("Average distance between modified frame_end from the first file and frame_end from the second file:", average_diff)


# Calculate the difference again for clarity
diff = modified_frame_end_df1 - df2['frame_start']

# Assuming differences are integers and you've shifted them to be non-negative if necessary
# For example, if you have negative values, you could shift all values by subtracting the min(diff)
if diff.min() < 0:
    diff_adjusted = diff - diff.min()
else:
    diff_adjusted = diff

# Now, use np.bincount to count occurrences of each difference
# Note: diff_adjusted needs to be cast to integers if not already
diff_counts = np.bincount(diff_adjusted.astype(int))

# To display the counts for each difference
for difference, count in enumerate(diff_counts):
    print(f"Difference: {difference + diff.min()}, Count: {count}")
```
